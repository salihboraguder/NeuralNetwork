```python
import numpy as np
```

```python
def sigmoid(x):
    return 1/(1+np.exp(x))
```

```python
class Dense:
    def __init__(self,activation,units):
        self.units = units
        self.activation = activation
    def relu(self,x):
        return max(0,x)

    def linear(self,x):
        return x

    def sigmoid(self,x):
        return 1/(1+np.exp(x))
    
    def forward(self,X):
        features = X.shape[1]
        X = X.T
        #weight and bias array initialization
        b = (np.ones(self.units) * 0.01).reshape(self.units,1)

        W = np.random.rand(self.units,features)
        
        
        result = np.dot(W,X) + b
        if self.activation =="relu":
            result = self.relu(result)
        elif self.activation =="linear":
            result = self.linear(result)
        else:
            
            result = self.sigmoid(result)
        return W,result,b
```

```python
a = Dense("sigmoid",units =10)
```

```python
X = np.random.rand(150,5)
```

```python
weights, res,b_arr= a.forward(X)
```
